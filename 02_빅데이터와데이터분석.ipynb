### 2.1 데이터 분석과 머신러닝을 위한 강력한 프로그래밍 언어 : 파이썬

```python
animals = ['lion', 'tiger', 'cat', 'dog']
animals.sort() # animals 리스트 내부 문자열을 알파벳 순으로 정렬 
animals
```

```python
animals.append('rabbit') # animals 리스트에 새 원소를 추가
animals
```

```python
animals.reverse() # animals 리스트를 원래 원소의 역순으로 재배열
animals
```

### 2.2 모듈의 개념과 활용

```python
def max2(a, b):
    if a > b:
        return a
    else:
        return b

max2(100, 200)
```

```python
import math # 수학 함수가 있는 모듈을 코드 안으로 가져온다
a = math.fabs(-10) # -10의 절대값을 구하는 함수 
print('math.fabs(-10) :', a)
b = math.sin(math.pi/2.0) # sin() 90도를 구하는 함수(라디안 각도 사용) 
print('math.sin(m.pi/2.0) :', b)
```
